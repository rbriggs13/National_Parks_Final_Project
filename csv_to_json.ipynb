```python
import pandas as pd
import json
```

```python
parks_df = pd.read_csv('Resources/national_parks.csv')
parks_df.head()
```

```python
parks_df = parks_df.set_index('Park Name')
parks_df
```

```python
species_df = pd.read_csv('Resources/clean_species.csv')
species_df.head()
```

```python
parks_data = parks_df.to_json(orient='index')
parsed = json.loads(parks_data)
parks_data = json.dumps(parsed, indent=4)
```

```python
with open("Dashboard/Resources/data/parks_data.json", "w") as outfile:
    outfile.write(parks_data)
```

```python
species_data = species_df.to_json(orient='records')
parsed = json.loads(species_data)
species_data = json.dumps(parsed, indent=4)
```

```python
with open("Dashboard/Resources/data/species_data.json", "w") as outfile:
    outfile.write(species_data)
```

```python
table_df = species_df.drop(columns = ['Species ID', 'Park Code', 'Order', 'Family', 'Occurrence'])
table_df.head()
```

```python
table_df = table_df[['Common Names', 'Scientific Name', 'Category', 'Nativeness', 'Conservation Status', 'Park Name']]
table_df.head()
```

```python
table_df.groupby(['Park Name'])['Scientific Name'].count()
```

```python
table_data = table_df.to_json(orient='records')
parsed = json.loads(table_data)
table_data = json.dumps(parsed, indent=4)
```

```python
with open("Dashboard/Resources/data/table_data.json", "w") as outfile:
    outfile.write(table_data)
```

```python
category_list = table_df['Category'].unique().tolist()
```

```python
category_data = json.dumps(category_list)
category_data
```

```python
with open("Dashboard/Resources/data/category_data.json", "w") as outfile:
    outfile.write(category_data)
```

```python
#import dependencies
import pandas as pd
```

```python
#read in csv files to make dataframes
parks_df = pd.read_csv('Resources/parks.csv')
visit_df = pd.read_csv('Resources/visitors.csv')
parks_df.head()
```

```python
visit_df.head()
```

```python
#rename the columns
visit_df = visit_df.rename(columns = {'Title':'Park Name', 'ParkType':'Rank', 'Bookmark': 'Visitors', 'Unnamed: 3': 'Percentage'})
```

```python
#clean the visitors dataframe
visit_df = visit_df.drop([0,1,2])
visit_df = visit_df.reset_index()
visit_df = visit_df.drop(columns = 'index')
visit_df.head()
```

```python
#Remove the NP from the Park Names in the visit_df
counter = 0

for name in visit_df['Park Name']:

    new_name = name.replace(' NP', '')

    visit_df['Park Name'][counter] = new_name
    counter += 1


visit_df.head(30)
```

```python
#Remove the & PRES from the rows that contain it
edit_df = visit_df[visit_df['Park Name'].str.contains(' & PRES')]

counter = 0
index= edit_df.index

for name in edit_df['Park Name']:

    new_name = name.replace(' & PRES', '')

    edit_df['Park Name'][index[counter]] = new_name
    counter += 1

    
gates = edit_df.loc[62,'Park Name']
st = edit_df.loc[56, 'Park Name']

print(gates)
edit_df.head(10)
```

```python
#return the new names to the original dataframe

replace = edit_df['Park Name'].tolist()

visit_df.loc[index, 'Park Name'] = replace
visit_df.head(30)
```

```python
#remove the phrase national park from the park names in the park_df
counter = 0

for name in parks_df['Park Name']:
    
    new_name = name.replace(' National Park', '')
    
    parks_df['Park Name'][counter] = new_name
    counter += 1

parks_df.head(20)
```

```python
#remove the 'and preserve' from the names
edit_df = parks_df[parks_df['Park Name'].str.contains(' and Preserve')]

counter = 0
index= edit_df.index

for name in edit_df['Park Name']:

    new_name = name.replace(' and Preserve', '')

    edit_df['Park Name'][index[counter]] = new_name
    counter += 1

edit_df.head(10)
```

```python
#fix the remaining names for issues
edit_df.loc[18, 'Park Name'] = gates
edit_df.loc[52, 'Park Name'] = st

edit_df.head(10)
```

```python
#add the values back the original dataframe
replace = edit_df['Park Name'].tolist()

parks_df.loc[index, 'Park Name'] = replace
parks_df.head(20)
```

```python
#merge the data frames
national_parks_df = pd.merge(parks_df, visit_df, on='Park Name', how='left')
national_parks_df.head()
```

```python
national_parks_df.isna().sum()
```

```python
national_parks_df = national_parks_df.dropna()
national_parks_df.isna().sum()
```

```python
#write the cleaned file to a new csv file
national_parks_df.to_csv('Resources/national_parks.csv', index=False)
```

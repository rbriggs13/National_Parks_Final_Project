```python
import pandas as pd
```

```python
#read in csv files to make dataframes
species_df = pd.read_csv('Resources/clean_species.csv')
natparks_df = pd.read_csv('Resources/national_parks.csv')
species_df.head()
```

```python
natparks_df.head()
```

```python
#merge the data frames
parkspecies_df = pd.merge(natparks_df, species_df, on='Park Name', how='left')
parkspecies_df.head(200)
```

```python
parkspecies_df.isna().sum()
```

```python
#write the cleaned file to a new csv file
parkspecies_df.to_csv('Resources/parkspecies.csv', index=False)
```

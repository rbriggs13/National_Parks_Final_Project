```python
#import dependencies
import pandas as pd
```

```python
species_df = pd.read_csv('Resources/species.csv')
species_df.head()
```

```python
#exploring the data
species_df.groupby(['Park Name'])['Scientific Name'].count()
```

```python
species_df.groupby(['Scientific Name'])['Park Name'].count().sort_values(ascending=False)
```

```python
species_df['Occurrence'].value_counts()
```

```python
species_df['Nativeness'].value_counts()
```

```python
#cleaning the data frame
clean_species_df = species_df[(species_df['Occurrence'].isin(['Present', 'Not Confirmed'])) & (species_df['Nativeness'] != 'Unknown')]
clean_species_df['Occurrence'].value_counts()
```

```python
clean_species_df['Nativeness'].value_counts()
```

```python
clean_species_df.head()
```

```python
clean_species_df = clean_species_df.drop(columns = ['Record Status', 'Abundance', 'Seasonality', 'Unnamed: 13'])
clean_species_df.head(10)
```

```python
clean_species_df['Conservation Status'].value_counts()
```

```python
clean_species_df['Conservation Status'].isnull().sum()
```

```python
clean_species_df['Conservation Status'] = clean_species_df['Conservation Status'].fillna(value = 'Least Concern')
clean_species_df['Conservation Status'].value_counts()
```

```python
clean_species_df = clean_species_df[(clean_species_df['Conservation Status'] != 'Under Review')]
clean_species_df['Conservation Status'].value_counts()
```

```python
clean_species_df = clean_species_df[(clean_species_df['Conservation Status'] != 'Proposed Endangered')]
clean_species_df['Conservation Status'].value_counts()
```

```python
clean_species_df = clean_species_df[(clean_species_df['Conservation Status'] != 'Proposed Threatened')]
clean_species_df['Conservation Status'].value_counts()
```

```python
clean_species_df.head()
```

```python
clean_species_df.isnull().sum()
```

```python
clean_species_df = clean_species_df.dropna()
clean_species_df.isnull().sum()
```

```python
clean_species_df.head()
```

```python
clean_species_df = clean_species_df.reset_index()
clean_species_df = clean_species_df.drop(columns = 'index')
clean_species_df.head()
```

```python
#add a column for park ids
park_ids = []

for species in clean_species_df['Species ID']:

    code = species.split('-', 1)
    park = code[0]
    park_ids.append(park)
```

```python
clean_species_df['Park Code'] = park_ids
clean_species_df.head()
```

```python
#remove the phrase national park from the park names in the park_df
counter = 0

for name in clean_species_df['Park Name']:
    
    new_name = name.replace(' National Park', '')
    
    clean_species_df['Park Name'][counter] = new_name
    counter += 1

clean_species_df.head()
```

```python
#remove the 'and preserve' from the names
edit_df = clean_species_df[clean_species_df['Park Name'].str.contains(' and Preserve')]

counter = 0
index= edit_df.index

for name in edit_df['Park Name']:

    new_name = name.replace(' and Preserve', '')

    edit_df['Park Name'][index[counter]] = new_name
    counter += 1

edit_df.head(10)
```

```python
#fix the gates of the artic name
counter = 0
index= edit_df.index

for name in edit_df['Park Name']:

    new_name = name.replace(' Of The', ' of the')

    edit_df['Park Name'][index[counter]] = new_name
    counter += 1

edit_df.head(10)
```

```python
#fix the st. elias name
counter = 0
index= edit_df.index

for name in edit_df['Park Name']:

    new_name = name.replace(' - St', '-St.')

    edit_df['Park Name'][index[counter]] = new_name
    counter += 1

edit_df.head(10)
```

```python
#add the values back the original dataframe
replace = edit_df['Park Name'].tolist()

clean_species_df.loc[index, 'Park Name'] = replace
clean_species_df.head(20)
```

```python
clean_species_df = clean_species_df.replace('"',"'")
clean_species_df
```

```python
#export to csv file
clean_species_df.to_csv('Resources/clean_species.csv', index = False)
```

```python
#clean the dataframe for machine learning
learning_species_df = clean_species_df.drop(columns = ['Species ID', 'Order', 'Family', 'Common Names', 'Occurrence', 'Park Code'])
learning_species_df.head()
```

```python
#removing the species from the scientific name but not the genus
genus_list = []

for species in learning_species_df['Scientific Name']:

    species_split = species.split(' ', 1)
    genus = species_split[0]
    genus_list.append(genus)


genus_list
```

```python
learning_species_df['Genus'] = genus_list
learning_species_df.head()
```

```python
learning_species_df = learning_species_df[['Genus', 'Category', 'Park Name', 'Conservation Status', 'Nativeness', 'Scientific Name']]
learning_species_df.head()
```

```python
learning_species_df = learning_species_df.drop(columns = 'Scientific Name')
learning_species_df.head()
```

```python
learning_species_df.to_csv('Resources/learning_species.csv', index = False)
```
